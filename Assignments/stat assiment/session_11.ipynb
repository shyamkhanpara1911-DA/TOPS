```python
# Task 1 – Simulate CLT in Python 
import numpy as np
import matplotlib.pyplot as plt

# Skewed dataset (Instagram likes)
likes = np.random.exponential(scale=100, size=10000)

sample_means = []

# Take 1000 samples of size 30
for i in range(1000):
    sample = np.random.choice(likes, size=30)
    sample_means.append(np.mean(sample))

# Plot histogram
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem Simulation (Sample Size = 30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

# Task 2 – Explanation 
Answer:

Even if the original data is not normally distributed, the average of many random samples tends to follow a normal distribution when the sample size is large enough.

Example (Zomato):

Suppose Zomato order amounts are highly irregular because some customers place small orders while others place very large orders. If we repeatedly take random samples of 30 orders and calculate the average order amount, the distribution of these sample averages becomes approximately normal. This helps analysts make reliable predictions and estimates.

# Task 3 – Real-World Example

Flipkart Example:

Flipkart wants to estimate the average delivery time across India. Instead of checking millions of deliveries, the company randomly selects many samples of 30 deliveries each and calculates their average delivery times. According to the Central Limit Theorem, these sample means form an approximately normal distribution, allowing Flipkart to estimate the overall average delivery time and build confidence intervals for better business decisions.

```python
#Task 4 – Compare Sample Size
import numpy as np
import matplotlib.pyplot as plt

likes = np.random.exponential(scale=100, size=10000)

means_5 = []
means_30 = []

# Sample size = 5
for i in range(1000):
    sample = np.random.choice(likes, size=5)
    means_5.append(np.mean(sample))

# Sample size = 30
for i in range(1000):
    sample = np.random.choice(likes, size=30)
    means_30.append(np.mean(sample))

# Histogram for n = 5
plt.figure(figsize=(8,4))
plt.hist(means_5, bins=30, edgecolor='black')
plt.title("Sample Size = 5")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# Histogram for n = 30
plt.figure(figsize=(8,4))
plt.hist(means_30, bins=30, edgecolor='black')
plt.title("Sample Size = 30")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

Generate Python code that demonstrates the Central Limit Theorem using random samples from an exponential distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=2, size=10000)

sample_means = []

for i in range(1000):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT using Exponential Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
