```python
# task 1 Random Sampling
import random

user_ids = list(range(1, 201))

sample = random.sample(user_ids, 20)

print(sample)
```

```python
# task 2 Stratified Sampling
import pandas as pd

cities = ["Ahmedabad"]*500 + ["Surat"]*300 + ["Vadodara"]*200

df = pd.DataFrame({"City": cities})

sample = df.groupby("City", group_keys=False).apply(
    lambda x: x.sample(frac=60/1000)
)

print(sample)
```

```python
# task 3 Systematic Sampling 
reviews = list(range(1,501))

sample = reviews[::10]

print(sample)
```

```python
# task 5    IPL Stratified Sampling 
import pandas as pd

teams = ["CSK"]*40 + ["MI"]*30 + ["RCB"]*20 + ["GT"]*10

df = pd.DataFrame({"Team":teams})

sample = df.groupby("Team",group_keys=False).apply(
    lambda x:x.sample(frac=40/100)
)

print(sample)
```
