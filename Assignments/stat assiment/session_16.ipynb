```python
# task 1 (Songs Data)
import pandas as pd
songs = pd.DataFrame({
    "danceability":[0.75,0.60,0.82,0.55,0.90,0.72,0.68,0.50,0.85,0.77],
    "popularity":[85,70,92,60,98,80,75,55,95,88]
})

pearson_corr = songs["danceability"].corr(
    songs["popularity"]
)

print("Pearson Correlation:", pearson_corr)
```

```python
#Task 2 (Spearman)
from scipy.stats import spearmanr
runs = [8000,6500,5200,2500,4200,3500,2400,3000]
followers = [270,35,48,15,18,12,30,14]

corr, p = spearmanr(runs, followers)

print("Spearman Correlation:", corr)
print("P-value:", p)
```

```python
#Task 3 (Correlation Matrix)
products = pd.DataFrame({
    "price":[55000,25000,3000,5000,20000,45000,4000,15000,1200,800],
    "user_rating":[4.5,4.3,4.2,4.4,4.1,4.6,4.0,4.5,4.2,4.1],
    "num_reviews":[1200,2500,1800,1500,900,700,1300,850,600,500]
})

corr_matrix = products.corr()
corr_matrix
```

```python
#task 5 Heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

# Task 5 (Theory)
Correlation does not imply causation.
Hidden variables can create misleading correlations.
