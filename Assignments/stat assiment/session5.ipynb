```python
# Task 1 – Percentiles 
import numpy as np

delivery_times = [18, 22, 25, 20, 30, 28, 24, 19, 35, 26,
                  27, 21, 23, 29, 31, 17, 32, 34, 20, 25]

p25 = np.percentile(delivery_times, 25)
p50 = np.percentile(delivery_times, 50)
p75 = np.percentile(delivery_times, 75)

print("25th Percentile:", p25)
print("50th Percentile (Median):", p50)
print("75th Percentile:", p75)
```

```python
# Task 2 – Quartiles 
import numpy as np

scores = [45, 60, 78, 90, 55, 68, 85, 72, 95, 80]

Q1 = np.quantile(scores, 0.25)
Q2 = np.quantile(scores, 0.50)
Q3 = np.quantile(scores, 0.75)

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
```

```python
 # Task 3 – IQR 
 import numpy as np

steps = [6500, 7000, 7200, 6800, 7500, 8000, 8200, 7700,
         6900, 7100, 7300, 7600, 8100, 8400, 7900,
         7800, 7000, 6900, 7200, 7500, 7700, 8000,
         8100, 8300, 8500, 7600, 7400, 7100, 6900, 7000]

Q1 = np.percentile(steps, 25)
Q3 = np.percentile(steps, 75)

IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
```

```python
# Task 4 – Outlier Detection 
import numpy as np

orders = [200, 250, 300, 275, 260, 290, 310, 220, 240, 1500]

Q1 = np.percentile(orders, 25)
Q3 = np.percentile(orders, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = []

for value in orders:
    if value < lower or value > upper:
        outliers.append(value)

print("Outliers:", outliers)
```
