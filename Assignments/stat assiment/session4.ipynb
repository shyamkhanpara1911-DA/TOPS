```python
#task 2
import statistics

orders = [250, 400, 350, 300, 500]

variance = statistics.variance(orders)
std_dev = statistics.stdev(orders)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
#task 3
import statistics

data_usage = [3.2, 2.8, 3.5, 3.0, 2.9, 3.1]

mean = statistics.mean(data_usage)
std_dev = statistics.stdev(data_usage)
cv = (std_dev / mean) * 100

print("Mean:", mean)
print("Standard Deviation:", std_dev)
print("Coefficient of Variation:", cv)
```
