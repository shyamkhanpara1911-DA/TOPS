```python
#Task 1 – One-Sample t-test
from scipy.stats import ttest_1samp

# Sample data (delivery times in minutes)
delivery_times = [28, 32, 27, 35, 29, 31, 30, 33, 26, 34]

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(delivery_times, 30)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Decision
if p_value < 0.05:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
#task 2 – Independent (Two-Sample) t-test
from scipy.stats import ttest_ind

# Before installing Google Fit
before = [4200, 4500, 4800, 5000, 4700, 4600, 4900, 5100, 4400, 4300]

# After installing Google Fit
after = [6200, 6500, 6800, 7000, 6700, 6600, 6900, 7100, 6400, 6300]

t_stat, p_value = ttest_ind(before, after)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Average steps increased significantly.")
else:
    print("No significant increase.")
```

```python
#Task 3 – Paired t-test
from scipy.stats import ttest_rel

# Screen time before Focus Mode
before = [6.5, 7.0, 5.8, 6.2, 7.5, 6.8, 7.2, 6.0, 6.7, 7.1]

# Screen time after Focus Mode
after = [5.2, 6.0, 5.0, 5.5, 6.3, 5.8, 6.1, 5.3, 5.7, 6.2]

t_stat, p_value = ttest_rel(before, after)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Focus Mode significantly reduced screen time.")
else:
    print("No significant reduction.")
```

```python
#Task 4 – Compare Zomato vs Swiggy Ratings
from scipy.stats import ttest_ind

zomato = [4.5, 4.2, 4.0, 4.3, 4.6, 4.4, 4.1, 4.3, 4.5, 4.2,
           4.4, 4.3, 4.1, 4.5, 4.6]

swiggy = [4.1, 4.0, 3.9, 4.2, 4.0, 4.1, 3.8, 4.0, 4.2, 4.1,
           3.9, 4.0, 4.1, 4.2, 4.0]

t_stat, p_value = ttest_ind(zomato, swiggy)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("The average ratings are significantly different.")
else:
    print("No significant difference in ratings.")
```

Why this test?

An independent two-sample t-test is used because the ratings come from two different groups of users (Zomato users and Swiggy users). The samples are independent of each other.

```python
#Task 5 – One-Sample t-test
from scipy.stats import ttest_1samp

upi_transactions = [18, 20, 19, 22, 21, 20, 23, 19, 18, 21]

t_stat, p_value = ttest_1samp(upi_transactions, 20)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```
