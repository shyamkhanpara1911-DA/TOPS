```python
#Task 1
# Generate 1000 Normal Random Numbers & Plot

import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(500, 100, 1000)

plt.hist(data, bins=30)

plt.title("Normal Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
# Task 2 Z-Score
x = 9500
mean = 8000
std = 1200

z = (x - mean) / std

print("Z-Score =", z)
```

```python
#Task 3 Standard Normal Curve
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4,4,500)

y = norm.pdf(x)

plt.plot(x,y)

plt.fill_between(x,y,where=(x>=-1)&(x<=1))

plt.title("Standard Normal Distribution")

plt.show()
```

```python
#Task 4 Probability Less Than 25 Minutes
from scipy.stats import norm

probability = norm.cdf(25,loc=30,scale=5)

print(probability)
```
