```python
# Task 1 – Bernoulli Trial 
import numpy as np

p = 0.85

results = np.random.binomial(1, p, 20)

print(results)
```

```python
#Task 2 – Binomial Distribution
from scipy.stats import binom

n = 5
k = 3
p = 0.9

probability = binom.pmf(k, n, p)

print("Probability =", probability)
```

```python
# Task 3 – Poisson Distribution
from scipy.stats import poisson

average = 4
k = 6

probability = poisson.pmf(k, average)

print("Probability =", probability)
```

```python
#Task 4 – Poisson Simulation + Plot
import numpy as np
import matplotlib.pyplot as plt

reviews = np.random.poisson(10, 12)

plt.plot(reviews, marker='o')

plt.title("Flipkart Product Reviews")
plt.xlabel("Hour")
plt.ylabel("Reviews")

plt.show()
```
