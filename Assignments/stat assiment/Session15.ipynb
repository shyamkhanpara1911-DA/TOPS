```python
import pandas as pd
from scipy.stats import f_oneway

df = pd.read_csv("purchase_data.csv")

g1 = df[df["age_group"]=="18-25"]["purchase_value"]
g2 = df[df["age_group"]=="26-35"]["purchase_value"]
g3 = df[df["age_group"]=="36-50"]["purchase_value"]

f_stat, p_value = f_oneway(g1, g2, g3)

print("F-statistic:", round(f_stat,4))
print("P-value:", round(p_value,4))

if p_value < 0.05:
    print("Reject the Null Hypothesis")
    print("There is a significant difference in average purchase values.")
else:
    print("Fail to Reject the Null Hypothesis")
    print("There is no significant difference in average purchase values.")
```
